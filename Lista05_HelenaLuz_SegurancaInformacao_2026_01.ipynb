```python
DEFAULT_BLOWFISH_KEY = bytes([65, 66, 67, 68, 69])

DEFAULT_BLOWFISH_PADDING_STYLE = 'pkcs7'
```

```python
def quebrarPorBloco(ciphertext):
  block_size = Blowfish.block_size
  for i in range(0, len(ciphertext), block_size):
      block = ciphertext[i:i+block_size]
      formatted_block = " ".join([f"{byte:02X}" for byte in block])
      print(formatted_block)
```

```python
def criptografarECB(plaintext):
    plaintext_bytes = plaintext.encode('utf-8')
    cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_ECB)
    padded_plaintext = pad(plaintext_bytes, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
    ciphertext = cipher.encrypt(padded_plaintext)

    print(f"{plaintext}:")
    quebrarPorBloco(ciphertext)

    return ciphertext
```

```python
def criptografarCTR(plaintext, nonce_bytes):
    plaintext_bytes = plaintext.encode('utf-8')
    cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_CTR, nonce=nonce_bytes)
    padded_plaintext = pad(plaintext_bytes, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
    ciphertext = cipher.encrypt(padded_plaintext)

    print(f"{plaintext} (CTR com Nonce={list(nonce_bytes)}):")
    quebrarPorBloco(ciphertext)
```

```python
def getExtensao(ciphertext):
  ciphertext_length_bytes = len(ciphertext)

  print(f"Extensão do texto cifrado: {ciphertext_length_bytes} bytes")
```

```python
def criptografarCBC(plaintext, iv_bytes):
    plaintext_bytes = plaintext.encode('utf-8')
    cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_CBC, iv=iv_bytes)
    padded_plaintext = pad(plaintext_bytes, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
    ciphertext = cipher.encrypt(padded_plaintext)

    print(f"{plaintext} (CBC com IV={list(iv_bytes)}):")
    quebrarPorBloco(ciphertext)

    return ciphertext
```

```python
def descriptografarCBC(ciphertext, iv_bytes):
    cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_CBC, iv=iv_bytes)
    decrypted_padded_text = cipher.decrypt(ciphertext)
    plaintext_bytes = unpad(decrypted_padded_text, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
    plaintext = plaintext_bytes.decode('utf-8')
    return plaintext
```

```python
def descriptografarECB(ciphertext, key_bytes):
    cipher = Blowfish.new(key_bytes, Blowfish.MODE_ECB)
    decrypted_padded_text = cipher.decrypt(ciphertext)
    plaintext_bytes = unpad(decrypted_padded_text, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
    plaintext = plaintext_bytes.decode('utf-8')
    return plaintext
```

```python
def xor_bytes(a, b):
    return bytes(x ^ y for x, y in zip(a, b))
```

```python
def criptografarPCBC(plaintext, iv):
    plaintext_bytes = plaintext.encode('utf-8')
    bloco = Blowfish.block_size
    cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_ECB)

    plaintext_padded = pad(plaintext_bytes, bloco)

    ciphertext = b""
    prev_plain = iv
    prev_cipher = iv

    for i in range(0, len(plaintext_padded), bloco):
        bloco_plain = plaintext_padded[i:i+bloco]

        entrada = xor_bytes(bloco_plain, xor_bytes(prev_plain, prev_cipher))

        bloco_cipher = cipher.encrypt(entrada)
        ciphertext += bloco_cipher

        prev_plain = bloco_plain
        prev_cipher = bloco_cipher

    quebrarPorBloco(ciphertext)
```

```python
getExtensao(cypherText)
```

# Caso 4 - Criptografe a sequência de bytes [8, 8, 8, 8, 8, 8, 8, 8] utilizando o modo de operação ECB.

4.1. Qual o conteúdo do texto cifrado?

```python
cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_ECB)
plaintext = bytes([8, 8, 8, 8, 8, 8, 8, 8])
padded_plaintext = pad(plaintext, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)

ciphertext= cipher.encrypt(padded_plaintext)

print(f"[8, 8, 8, 8, 8, 8, 8, 8]:\n{ciphertext.strip()}")
```

4.2. Compare os primeiros 8 bytes do texto cifrado com o último bloco cifrado da questão anterior. Que conclusão é
possível obter?

```python
cipher = Blowfish.new(DEFAULT_BLOWFISH_KEY, Blowfish.MODE_ECB)
plaintext_case4 = bytes([8, 8, 8, 8, 8, 8, 8, 8])
padded_plaintext_case4 = pad(plaintext_case4, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)
ciphertext_case4 = cipher.encrypt(padded_plaintext_case4)

last_block_case3 = cypherText_caso3[8:]
first_block_case4 = ciphertext_case4[:8]

print(f"Último bloco cifrado de 'SABONETE': {last_block_case3}")
print(f"Primeiro bloco cifrado de '[8, 8, 8, 8, 8, 8, 8, 8]' : {first_block_case4}")
```

Os primeiros 8 bytes do texto cifrado do Caso 4 são idênticos ao último bloco cifrado do Caso 3.
Isso acontece porque, no modo de operação ECB (Electronic Codebook), blocos de plaintext idênticos (neste caso, o bloco de preenchimento ou o byte 8 repetido 8 vezes) sempre resultarão em blocos de ciphertext idênticos, desde que a chave de criptografia seja a mesma. O Blowfish tem um tamanho de bloco de 8 bytes.

# Caso 10 - Utilizando a chave “ABCDE”, criptografe o arquivo PDF que foi publicado para este exercício, salvando-o em disco com o nome “saída.bin”. Utilize o modo de operação ECB.

```python
chave_caso10 = b"ABCDE"

pdf_file_name = "LE - Criptografia Blowfish (v4).pdf"

try:
    with open(pdf_file_name, 'rb') as f:
        plaintext_pdf = f.read()

    cipher = Blowfish.new(chave_caso10, Blowfish.MODE_ECB)
    padded_plaintext_pdf = pad(plaintext_pdf, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)

    ciphertext_pdf = cipher.encrypt(padded_plaintext_pdf)

    output_file_name = "saida.bin"
    with open(output_file_name, 'wb') as f:
        f.write(ciphertext_pdf)

    print(f"Arquivo '{pdf_file_name}' criptografado com sucesso para '{output_file_name}' usando ECB.")

except FileNotFoundError:
    print(f"Erro: O arquivo '{pdf_file_name}' não foi encontrado. Por favor, verifique o caminho e o nome do arquivo e certifique-se de que ele foi enviado para o Colab.")
except Exception as e:
    print(f"Ocorreu um erro durante a criptografia: {e}")
```

10.1. Qual o tamanho em bytes dos dois arquivos?


```python
print(f"Tamanho do arquivo original: {len(plaintext_pdf)} bytes")
print(f"Tamanho do arquivo cifrado: {len(ciphertext_pdf)} bytes")
```

# Caso 11 - Descriptografe o arquivo saida.bin, que você gerou anteriormente. Utilize a chave “ABCDE” e o modo de operação ECB. Salve o arquivo com o nome “descriptografado.pdf”. Tente abrir o arquivo.


```python
try:
    with open('saida.bin', 'rb') as f:
        ciphertext_pdf = f.read()

    decrypted_padded_text = cipher.decrypt(ciphertext_pdf)
    plaintext_bytes_for_pdf = unpad(decrypted_padded_text, Blowfish.block_size, style=DEFAULT_BLOWFISH_PADDING_STYLE)

    output_decrypted_file_name = "descriptografado.pdf"
    with open(output_decrypted_file_name, 'wb') as f:
        f.write(plaintext_bytes_for_pdf)

    print(f"Arquivo '{output_decrypted_file_name}' salvo com sucesso.")

except FileNotFoundError:
    print("Erro: O arquivo 'saida.bin' não foi encontrado. Certifique-se de que a etapa de criptografia foi executada com sucesso.")
except Exception as e:
    print(f"Ocorreu um erro durante a descriptografia: {e}")
```
